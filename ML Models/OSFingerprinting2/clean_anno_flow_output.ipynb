```python
conditions = [
    (df['HTTP_Host_OS'] == 'Windows'),
    (df['HTTP_Host_OS'] == 'Mac OS X'),
    (df['HTTP_Host_OS'] == 'Android'),
    (df['HTTP_Host_OS'] == 'iOS'),
    (df['HTTP_Host_OS'] == 'Ubuntu'),
    (df['HTTP_Host_OS'] == 'Linux'),
    (df['HTTP_Host_OS'] == 'Darwin'),
    (df['HTTP_Host_OS'] == 'Windows Phone'),
    (df['HTTP_Host_OS'] == 'Fedora'),
    (df['HTTP_Host_OS'] == 'BlackBerry'),
    (df['HTTP_Host_OS'] == 'Debian'),
    (df['HTTP_Host_OS'] == 'Windows CE'),
    (df['HTTP_Host_OS'] == 0)

]

choices = ['Windows_PC', 'MAC', 'Mobile Phone', 'MAC', 'Linux_PC', 'Linux_PC', 'MAC_PC','MS_Phone','Linux_PC','Mobile Phone','Linux_PC', 'Windows_PC', 'Undetectable']

df2['Device_Category'] = np.select(conditions, choices, default='Undetectable')
df2.head(25)
```
